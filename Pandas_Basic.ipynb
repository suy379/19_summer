```python
#데이터 슬라이싱
##원하는 컬럼만 슬라이싱
df['Pclass'].head()
```

```python
##원하는 행만 슬라이싱
df[0:3]
```

```python
#iloc, loc
##iloc은 integer형으로만(위치 기반 인덱싱), loc은 컬럼네임으로만(명칭 기반 인덱싱) 슬라이싱 가능
data={'Name' : ['A','B','C','D'], 'Year' : [2011,2016,2017,2019], 'Gender' : ['M','F','M','M']}
data_df=pd.DataFrame(data, index=['one','two','three','four'])
data_df
```

```python
data_df.iloc[0,0]
```

```python
data_df.loc['three','Gender']
```

```python
data_df.loc['three':'four',:] #이 외에도, 스스로 여러 가지 슬라이싱을 해 보세요!
```

```python
rows=['one','four']
cols=['Year']
data_df.loc[rows,cols]
```

```python
#불린 인덱싱(Boolean Indexing): loc, iloc 슬라이싱보다 더 많이 써요
df[df.Pclass==3].head(7)
```

```python
##보고 싶은 칼럼만 떼서 보여주기
df[df.Age>60][['Name','Ticket']].head(7)
```

```python
#여러 조건을 중첩시켜 슬라이싱해 봅시다
df[(df.Age>60) & (df['Pclass']==1) & (df.Sex=='male')][['Name','Ticket']].head(7)
```

```python
con1=df.Age>60
con2=df['Pclass']==1
con3=df.Sex=='male'
df[con1 & con2 & con3][['Name','Ticket']].head(7)
```

```python
#정렬(Sorting) : sort_values()
titanic_sort=df.sort_values(by=['Name'],ascending=False) #ascending=True(오름차순), False(내림차순)
titanic_sort.head()
```

```python
#기본적 함수 적용(Aggregation 함수 적용) : min(),max(),sum(),mean(),count()...
df.count()
```

```python
#칼럼 지정
df[['Age','Fare']].sum()
```

```python
#그룹바이(Groupby): 정말정말 많이 써요!
titanic_group=df.groupby('Pclass')
titanic_group 
```

```python
##결과값 반환하려면 밑처럼 agg함수를 적용해줍시다
titanic_group=df.groupby('Pclass').mean()
titanic_group
```

```python
df.groupby('Pclass')['Age'].agg([max,min])
```

```python
##컬럼별로 적용하는 agg함수를 다르게 해봅시다
agg={'Age': 'mean', 'SibSp': 'sum', 'Fare': 'max'}
df.groupby('Pclass').agg(agg)
```

```python
#결측치 처리: 전처리의 기본!
##결측치(Missing data)란? 값이 없는, 즉 NULL 값을 의미합니다. NaN으로 표시
##결측치가 있는지 항상 전처리의 처음에 확인해야 하고, 이를 다른 값으로 대체해야 합니다.

#결측치 확인 : isna()
df.isna().head()
```

```python
df.isna().sum()
```

```python
#결측치 대체 : fillna()
df['Cabin']=df['Cabin'].fillna('C000')
df.head()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df.isna().sum()
```

```python
#결측치 삭제 : dropna()
##how 옵션엔 'all'(모든 컬럼에 NaN값이 있는 행 drop), 'any'(하나라도 NaN값이 있는 행 drop)
df=df.dropna(how='any')
df.isna().sum()
```
